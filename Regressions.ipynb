```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR

np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Polynomial Regression
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)

# Ridge Regression
ridge_reg = Ridge(alpha=1.0)
ridge_reg.fit(X, y)

# Lasso Regression
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X, y)

# Support Vector Regression
svr_reg = SVR(kernel='linear')
svr_reg.fit(X, y.ravel())

# Plotting the results
plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg.predict(X), color='red', label='Linear Regression')
plt.plot(X, poly_reg.predict(X_poly), color='green', label='Polynomial Regression')
plt.plot(X, ridge_reg.predict(X), color='purple', label='Ridge Regression')
plt.plot(X, lasso_reg.predict(X), color='orange', label='Lasso Regression')
plt.plot(X, svr_reg.predict(X), color='black', label='Support Vector Regression')
plt.legend()
plt.show()
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR

np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Polynomial Regression
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)

# Ridge Regression
ridge_reg = Ridge(alpha=1.0)
ridge_reg.fit(X, y)

# Lasso Regression
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X, y)

# Support Vector Regression
svr_reg = SVR(kernel='linear')
svr_reg.fit(X, y.ravel())

# Plotting the results
plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg.predict(X), color='red', label='Linear Regression')
plt.plot(X, poly_reg.predict(X_poly), color='green', label='Polynomial Regression')
plt.plot(X, ridge_reg.predict(X), color='purple', label='Ridge Regression')
plt.plot(X, lasso_reg.predict(X), color='orange', label='Lasso Regression')
plt.plot(X, svr_reg.predict(X), color='black', label='Support Vector Regression')
plt.legend()
plt.show()
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR

np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Polynomial Regression
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)

# Ridge Regression
ridge_reg = Ridge(alpha=1.0)
ridge_reg.fit(X, y)

# Lasso Regression
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(X, y)

# Support Vector Regression
svr_reg = SVR(kernel='linear')
svr_reg.fit(X, y.ravel())

# Plotting the results
plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg.predict(X), color='red', label='Linear Regression')
plt.plot(X, poly_reg.predict(X_poly), color='green', label='Polynomial Regression')
plt.plot(X, ridge_reg.predict(X), color='purple', label='Ridge Regression')
plt.plot(X, lasso_reg.predict(X), color='orange', label='Lasso Regression')
plt.plot(X, svr_reg.predict(X), color='black', label='Support Vector Regression')
plt.legend()
plt.show()
```
